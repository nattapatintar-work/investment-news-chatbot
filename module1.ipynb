```python
def greet_investor(name, portfolio_value):
    """
    This text between triple quotes is called a 'docstring'
    It explains what the function does — always write one!
    """
    message = f"Welcome back, {name}! Your portfolio is worth ${portfolio_value:,.2f}"
    return message    # return sends the result back to whoever called us
```

```python
def get_stock_price(ticker):
    """
    Fetch current stock price from a free API 
    We use Alpha Vantage - 25 free calls/day no credit card needed
    """ 
    API_KEY = "demo"
    url = f"https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={ticker}&apikey={API_KEY}"
    try:
        response = requests.get(url) # sends the HTTP request and store sever response (status code , headers ,body)
        data = response.json() # take the respone body and parse to Json 
        
        quota = data['Global Quote']
        price = quota['0.5 price']
        change = quota["10. change percent"]

        return {
            "ticker" : ticker ,
            "price" : float(price),
            "change_percent" : change
        }
    
    except Exception as error:
        print(f'Something went wrong : {error}')
        return None
    
result = get_stock_price('AAPL')
if result:
    print(f"Ticker: {result['ticker']}")
    print(f"Price:  ${result['price']}")
    print(f"Change: {result['change_percent']}")
```
